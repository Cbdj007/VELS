```python
import numpy as np
import pandas as pdpd
import matplotlib.pyplot as plt
import seaborn as sns

pizza_size=[[2],[3],[3.5],[4],[5],[2.5],[5.5],[4.5],[7],[6],[10],[9.5]]
pizza_price=[[100],[120],[150],[175],[200],[110],[220],[185],[400],[300],[700],[680]]

plt.figure
plt.title("Pizza prices")
plt.xlabel("pizza_size")
plt.ylabel("pizza_price")
plt.plot(pizza_size,pizza_price,'*')
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(pizza_size,pizza_price)
model.predict([[7.5]])
```
